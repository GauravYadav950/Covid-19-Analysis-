```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Covid 19 Data.csv')
df.head(5)
```

```python
df.shape
df.info()
df.isna().sum()
```

```python
df.describe(include='all')
```

```python
# Converting Date datatype to date

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.dtypes
```

```python
# Standardizing text Columns(Remove extra Space And convert names to tite case)
df['Country/Region'] = df['Country/Region'].str.strip().str.title()
```

```python
# Converting Numeric Columns into Integers

num_cols = ['Confirmed','Deaths','Recovered','Active']
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
df.dtypes
```

```python
# Cleaned  Text Columns 
text_cols = ['Province/State','Country/Region','WHO Region']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
    df[col] = df[col].replace({'Nan': None})   # convert 'nan' strings to actual NaN
```

```python
# Check Which Country Have how manys time Negative value

df[df['Active'] < 0].groupby('Country/Region').size()
```

```python
# Replacing Negative value to Zeroes 

df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
df['Active'] = df['Active'].apply(lambda x: max(x, 0))
df[df['Active'] < 0]
```

```python
df['WHO Region'] = df['WHO Region'].fillna('Unknown')
```

```python
df.rename(columns={
    'Country/Region': 'Country',
    'WHO Region': 'WHO_Region'
}, inplace=True)
df.head(10)
```

```python
df.drop(['Province/State','Lat','Long'], axis=1, inplace=True)
```
